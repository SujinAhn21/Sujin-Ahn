```python
# 0. 사용할  패키지  불러오기 
import numpy as np
from keras.models import Sequential 
from keras.layers import Dense 
import random
```

```python
# 1. 데이터셋  생성하기
x_train = np.random.random((1000, 1))
y_train = x_train * 2 + np.random.random((1000, 1)) / 3.0   
x_test = np.random.random((100, 1))
y_test = x_test * 2 + np.random.random((100, 1)) / 3.0                                                           
```

```python
import os
os.getcwd() 
```

```python
import pandas as pd
x_train=pd.read_csv('x_train.csv', sep=",", header=None)
y_train=pd.read_csv('y_train.csv', sep=",", header=None)
x_test=pd.read_csv('x_test.csv', sep=",", header=None)
y_test=pd.read_csv('y_test.csv', sep=",", header=None)
```

```python
x_train.describe()
```

```python
y_train.describe()
```

```python
x_test.describe()
```

```python
y_test.describe()
```

```python
# 2. 모델  구성하기
model = Sequential()
model.add(Dense(64, input_dim=4, activation='relu'))
model.add(Dense(64, activation='relu'))
model.add(Dense(1))
```

```python
# 3. 모델  학습과정  설정하기
model.compile(optimizer='rmsprop', loss='mse')
```

```python
# 4. 모델  학습시키기
hist = model.fit(x_train, y_train, epochs=10000, batch_size=64)
```

```python
# 5. 학습과정 살펴보기
%matplotlib inline
import matplotlib.pyplot as plt

plt.plot(hist.history['loss'])
plt.ylim(0.0, 0.5)
plt.ylabel('loss')
plt.xlabel('epoch')
plt.legend(['train'], loc='upper left')
plt.show()
```

```python
# 6. 모델  평가하기
loss = model.evaluate(x_test, y_test, batch_size=32) 
print('loss : ' , str(loss))
```
